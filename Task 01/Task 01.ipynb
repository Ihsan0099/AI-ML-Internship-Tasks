### 1: Importing important libraries 

```python
# Importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### 2: Loading DataSet

```python
# load the dataset 
df = pd.read_csv('Iris.csv')
# Display first few rows of the dataset
df.head()
```

### 3: Dataset Inspection

### 4: Data Visualization

Now we will use scatter plots, histograms, and box plots to understand relationships, distributions, and outliers.


### 4.1: scatter plots

```python
# Scatter plot to show relationships between SepalLength and SepalWidth
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', hue='Species', data=df)
```

```python
# Scatter plot to show relation b/w PetalLength and PetalWidth
sns.scatterplot(data=df, x='PetalLengthCm', y='PetalWidthCm', hue='Species')
```

```python
# Scatter plot to show relation b/w sepal lenght and petal length
sns.scatterplot(data = df, x='SepalLengthCm', y='PetalLengthCm', hue='Species')
```

```python
# scatter plot to show relation b/w sepal width and petal width
sns.scatterplot(data=df, x='SepalWidthCm', y='PetalWidthCm', hue="Species")
```

### 4.2: Histograms

```python
# histogram for numerical variables
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

sns.histplot(data=df, x="SepalLengthCm", kde=True, color="skyblue", ax=axs[0, 0])
sns.histplot(data=df, x="SepalWidthCm", kde=True, color="olive", ax=axs[0, 1])
sns.histplot(data=df, x="PetalLengthCm", kde=True, color="gold", ax=axs[1, 0])
sns.histplot(data=df, x="PetalWidthCm", kde=True, color="black", ax=axs[1, 1])

plt.tight_layout()
plt.show()

```

### 4.3: BoxPlots

```python
# boxplot to observe distribution of numerical values
fig, axs = plt.subplots(2, 2, figsize=(15, 8))
sns.boxplot(x=df['SepalLengthCm'], ax=axs[0, 0], color='skyblue')
axs[0, 0].set_title('Sepal Length')
axs[0, 0].set_xlabel('Sepal Length (cm)')
sns.boxplot(x=df['SepalWidthCm'], ax=axs[0, 1], color='lightgreen')
axs[0, 1].set_title('Sepal Width')
axs[0, 1].set_xlabel('Sepal Width (cm)')
sns.boxplot(x=df['PetalLengthCm'], ax=axs[1, 0], color='gold')
axs[1, 0].set_title('Petal Length')
axs[1, 0].set_xlabel('Petal Length (cm)')
sns.boxplot(x=df['PetalWidthCm'], ax=axs[1, 1], color='salmon')
axs[1, 1].set_title('Petal Width')
axs[1, 1].set_xlabel('Petal Width (cm)')
plt.tight_layout()
plt.show()
```
